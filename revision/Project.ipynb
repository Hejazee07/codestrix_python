```python
class User:
    def __init__(self,name,age,address):
        self.name=name
        self.age=age
        self.address=address
    def display_info(self):
        print(f"Thank You! {self.name.title()}")
        print(f"Your age is {self.age}")
        print(f"Your address is {self.address}")
```
