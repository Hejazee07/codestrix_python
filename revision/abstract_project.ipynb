```python
from abc import ABC,abstractmethod
```

```python
class IBank(ABC):
    def __init__(self):
        self.balance=10000
        self.transactions=[]
        self.pin=int(input("Enter your 4 digit PIN"))
    def deposit(self):
        try:
            amount=int(input("Enter your amount to be deposited: "))
            if amount<0:
                print("Invalid Input")
            self.balance+=amount
            self.transactions.append(f"Deposited Rs {amount}.")
            print(f"Amount deposited is {amount} and your new balance is {self.balance}.")
        except:
            print("Invalid Input.")
    def withdraw(self):
        try:
            amount=int(input("Enter the amount to be debited: "))
            entered=int(input("Enter 4 digit Pin"))
            if entered!=self.pin:
                print("Invalid Pin.")
                return
            if amount>self.balance:
                print("Insufficient Balance")
                return
            if not self.can_withdraw(amount):
                return
            self.balance-=amount
            self.transactions.append(f"Withdrawed Rs {amount}.")
            print(f"Amount withdrawed is {amount} and your new balance is {self.balance}.")
        except:
            print("Invalid Input.")
    def show_transactions(self):
        print("Transactions\n")
        if not self.transactions:
                print("No Transactions Yet.")
        else:
            for t in self.transactions:
                print(t)
    @abstractmethod
    def type_of_account(self):
         pass
    @abstractmethod
    def can_withdraw(self,amount):
        pass
```
